```python
#This code is saved as random_walk.py

from random import choice

class RandomWalk:
    """A class to generate random walks"""

    def __init__(self, num_points=5000):
        """Initialize attributes of a walk"""
        self.num_points = num_points

        #All walks start at (0,0)
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the wallk"""

        #Keep taking steps unitl the walk reaches the desired lenght
        while len(self.x_values) < self.num_points:

            #Decide which direction to go and how far to go in that direction
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance

            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance

            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position. We add the x_step/y_step values to the last value
            #stored in x_values/y_values lists at position -1 no matter how long the list is
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)
```

```python
import matplotlib.pyplot as plt

from random_walk import RandomWalk

#Make a random walk
while True:
    
    rw = RandomWalk()
    rw.fill_walk()

    #plot the points in the walk
    plt.style.use('classic')

    fig, ax = plt.subplots(figsize=(10, 6)) #sizes the screen
    
    point_numbers = range(rw.num_points) #generates a list of numbers equal to those of the walk
    
    #coloring the points along the walk
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15)
    
    #Emphasizaing the first and last points along the walk
    ax.scatter(0,0, c='green', edgecolors='none', s=100)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100)

    plt.show()
    
    #These lines of code will generate multiple walks
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```
